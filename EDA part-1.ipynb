```python
# Here we observed that,some of columns whoz caring int but shown in object dtype
# also some of columns having null values
```

```python
columns=['normalized-losses','bore','peak-rpm','horsepower','stroke','bore']
for i in columns:
    df[i]=pd.to_numeric(df[i])
```

```python
df.info()
```

```python
for i in columns:
    plt.figure(figsize=(10,8))
    sns.boxplot(df[i])
```

```python
#There are some features, which having some outliers
for i in columns:
    avg=df[i].median()
    df[i]=df[i].fillna(avg)
```

```python
df.isnull().sum()
```

```python
df[df['num-of-doors'].isnull()]
```

```python
df['num-of-doors'].value_counts()
```

```python
df['num-of-doors']=df['num-of-doors'].fillna("four")
```

```python
df.isnull().sum()
```

```python
df=df.dropna(axis=0)
```

```python
df.isnull().sum()
```

```python
df=df.reset_index(drop=True)
```

```python
df.info()
```

```python
df['symboling']=df['symboling'].astype("object")
```

```python
df['price']=df['price'].astype("int")
```
